```python
import pandas as pd
movies = pd.read_csv("tmdb_5000_movies.csv")
credits= pd.read_csv("tmdb_5000_credits.csv")
movies.head()
```

```python
credits.head()
```

```python
credits.columns = ['id', 'titles', 'cast', 'crew']
movies = movies.merge(credits, on='id')
movies.head()
```

```python
movies = movies[['title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head()
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        
        L.append(i['name'].replace(" ", "").lower())
    return L[:3]  

def get_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'director':
            return i['name'].replace(" ", "").lower()
    return ''

movies['cast'] = movies['cast'].apply(convert)
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['crew'] = movies['crew'].apply(get_director)
movies.rename(columns={'crew': 'director'}, inplace=True)

movies['overview'] = movies['overview'].fillna('')
movies['tags'] = movies['overview'] + " " + movies['genres'].apply(lambda x: " ".join(x)) + \
                 " " + movies['keywords'].apply(lambda x: " ".join(x)) + \
                 " " + movies['cast'].apply(lambda x: " ".join(x)) + \
                 " " + movies['director']
movies['tags'] = movies['tags'].str.lower()
```

```python
movies.head()
```

```python
from sentence_transformers import SentenceTransformer, util
import torch
model = SentenceTransformer('all-MiniLM-L6-v2')  


embeddings = model.encode(movies['tags'].tolist(), convert_to_tensor=True)

def recommend(movie_name):
    
    idx = movies[movies['title'] == movie_name].index[0]
    query_embedding = embeddings[idx]
    cos_scores = util.pytorch_cos_sim(query_embedding, embeddings)[0]
    top_results = torch.topk(cos_scores, k=6)
    
    for i in top_results.indices[1:]:  
        print(movies.iloc[i.item()]['title'])
```

```python
recommend("The Dark Knight")
```

```python
recommend("Avatar")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english')
vector = tfidf.fit_transform(movies['tags'])


similarity = cosine_similarity(vector)

# Recommend function
def recommend2(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = list(enumerate(similarity[index]))
    sorted_list = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]
    for i in sorted_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend2("Avatar")
```

```python
recommend2("The Dark Knight")
```
